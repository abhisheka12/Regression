```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
%matplotlib inline

df = pd.read_csv('')

df.head()

df.shape

df.info()

df.isnull()

df['Weight'] = pd.to_numeric(df['Weight'],errors = 'coerce')

df['Length'] = pd.to_numeric(df['Length'],errors = 'coerce')

df['Width'] = pd.to_numeric(df['Width'],errors = 'coerce')
df.info()

df['Weight'].fillna(df['Weight'].mean(), inplace=True)

df['Length'].fillna(df['Length'].mean(), inplace=True)

df['Width'].fillna(df['Width'].mean(), inplace=True)

df['Height'].fillna(df['Height'].mean(), inplace=True)

df.info()

df.describe()

df.value_counts()

### Graphs
sns.pairplot(df)

sns.boxplot(df['Height'])
# sns.boxplot(df['Cost'], df['Weight'])
# sns.boxplot(df['Cost'], df['Weight1'])
# sns.boxplot(df['Cost'], df['Length'])
# sns.boxplot(df['Cost'], df['Width'])

df_corr = df.corr()
sns.heatmap(df_corr, cmap='viridis', annot=True)

### model OLS

ols_statsmodel = sm.ols('Cost ~ Weight + Weight1 + Length + Height + Width',df).fit()
print(ols_statsmodel.params)
```
